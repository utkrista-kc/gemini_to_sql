```python
# Define the list of columns to keep
columns_to_keep = [
    'rank', 'finalWorth', 'category', 'personName', 'country',
    'city', 'source', 'industries', 'countryOfCitizenship', 'organization',
    'selfMade', 'status', 'gender', 'lastName', 'firstName',
    'title', 'birthYear', 
]

# Filter the DataFrame to include only the selected columns
filtered_raw_data = raw_data[columns_to_keep]

# Display the first few rows of the filtered DataFrame to verify

```

```python
filtered_raw_data.head()
```

```python
filtered_raw_data.columns
```

```python
# Define a dictionary for renaming columns dictionary
renaming_columns_dict = {
    'rank': 'rank',
    'finalWorth': 'final_worth',
    'category': 'category',
    'personName': 'person_name',
    'country': 'country',
    'city': 'city',
    'source': 'source',
    'industries': 'industries',
    'countryOfCitizenship': 'country_of_citizenship',
    'organization': 'organization',
    'selfMade': 'self_made',
    'status': 'status',
    'gender': 'gender',
    'lastName': 'last_name',
    'firstName': 'first_name',
    'title': 'title',
    'birthYear': 'birth_year'
}

# Rename the columns in the filtered DataFrame
filtered_raw_data_renamed = filtered_raw_data.rename(columns=renaming_columns_dict)

```

```python
# Display the first few rows of the renamed DataFrame to verify
filtered_raw_data_renamed.head()
```

```python
# Check for missing values in each column
missing_values = filtered_raw_data_renamed.isnull().sum()
print(missing_values)
```

```python
# Handling missing values
numerical_columns = ['birth_year']
filtered_raw_data_renamed[numerical_columns] = filtered_raw_data_renamed[numerical_columns].fillna(filtered_raw_data_renamed[numerical_columns].median())
```

```python
# Handling missing values
categorical_columns = ['country', 'city', 'organization', 'first_name', 'title']
for col in categorical_columns:
    filtered_raw_data_renamed[col] = filtered_raw_data_renamed[col].fillna('Unknown')
```

```python
missing_values_after_imputation = filtered_raw_data_renamed.isnull().sum()
print(missing_values_after_imputation)
```
